# Data Preprocess

```python
1. Remove highly correlated columns
2. Remove duplicate obsevations
3. Feature scaling
```

## 1. Remove highly correlated columns

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
warnings.simplefilter("ignore")

data = pd.read_excel("..\Data\Bank_Personal_Loan_Data.xlsx")
display(data.tail())
```

```python
data.set_index('ID', inplace=True)
data.drop('ZIP Code', axis=1, inplace=True)
display(data.head())
```

```python
# Observe correlation of features
import matplotlib.pyplot as plt
import seaborn as sns

cor = data.corr()
fig = plt.figure(figsize=(10, 10))
plt.title('Correlation')
ax = sns.heatmap(cor, annot=True, square=True)
```

```python
=> Here Age nad Experience are correlated to each other with 0.99
=> Lets drop experience from data
```

```python
data.drop('Experience', axis=1, inplace=True)
display(data.head())
```

```python
data[data.columns].corr()
```

```python
data.drop('CCAvg', axis=1, inplace=True)
display(data.head())
```

## 2. Remove duplicate obsevations

```python
x = data.shape
print(x)
data.drop_duplicates(inplace=True)
print(data.shape)

print(f'Number of duplicats: {x[0]-data.shape[0]}')
```

```python
=> 35 Duplicate observations removed.
```

## 3. Feature scaling

```python
1. Age and Income needs feature scaling
2. We know that age and income data distribution:
    Age    : 20 - 70
    Income : 10 - 220
3. Lets take Log Transform Scaling.
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# data['Age'] = np.log(data['Age'])
# data['Income'] = np.log(data['Income'])

sclr = StandardScaler()
data[['Age', 'Income', 'Family','Education']] = sclr.fit_transform(data[['Age', 'Income', 'Family','Education']])

display(data.head())
```

```python
print(min(data.Age), min( data.Income))
print(max(data.Age), max( data.Income))
```

```python
=> Great, its vary nice scaling.
```
